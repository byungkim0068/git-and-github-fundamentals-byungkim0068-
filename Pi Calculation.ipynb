```python
import numpy as np
import matplotlib.pyplot as plt

x = []
ndarts = []

for power in range(1, 7):
    n = 10 ** power
    ndarts.append(n)
    points = np.random.uniform(-1, 1, (2, n))
    inside_circle = np.sum(np.sum(points ** 2, axis=0) < 1)
    x.append(inside_circle / n * 4)

print(ndarts, x)
plt.plot(ndarts, x, 'o')
plt.xscale('log')  # Use a logarithmic scale for x-axis
plt.xlabel('Number of Darts')
plt.ylabel('Estimated Value of π')
plt.title('Monte Carlo Estimation of π')
plt.show()
```
